# Realizando consultas

```python
from connect import db
from bson.objectid import ObjectId

resultados = db.libros

# Consultar todo
for doc in resultados.find():
    print(doc)
```

```python
#Consultar por autor
for doc in resultados.find({'Autor':'Mario Vargas Llosa'}):
    print(doc['Titulo'],[doc['Anio']])
```

```python
#Consulta muestre libros mayores a el año 1960
for doc in resultados.find({'anio':{'$gt':1960}}):
    print(doc['titulo'],doc['anio'])
```

```python
#Consulta por genero literario 'Ficcion'
for doc in resultados.find({'generos':{'$gt':'ciencia ficción'}}):
    print(doc['titulo'],doc['anio'])
```

```python
#Libros con stock mayor a 2
for doc in resultados.find({'Stock':{'$gt':2}}):
    print(doc['Titulo'],doc['Anio'],doc['Stock'])
```

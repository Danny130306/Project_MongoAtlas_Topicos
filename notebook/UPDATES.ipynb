```python
from connect import db
from pymongo import UpdateOne

modificar = db.libros

# Actualizar stock de un libro
res = modificar.update_one(
    {'Titulo': 'La ciudad Verde'},
    {'$inc':
        {'Stock': 10}
    }
)

for doc in modificar.find({'Titulo':'La ciudad Verde'}):
    print(doc)


print('Cambios', res.matched_count, 'Total', res.modified_count)
```
